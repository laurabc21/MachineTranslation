```python
sentences = []
with open('wmtnews2013.en') as f:
  for line in f:
    sentences.append(line.strip())
del line, f
assert len(sentences) == 3000
```

```python
import google.generativeai as genai

genai.configure(api_key="##")

model = genai.GenerativeModel('gemini-2.0-flash')
```

```python
prompt = """Please translate the following English sentences into Catalan. Provide only the translations, without any additional explanations or comments:\n {}"""
```

```python
from time import sleep
from tqdm import tqdm
import json

batch_prompt = ""
batch_size = 25
orig_batch = []

with open('translations.json', 'w') as f:
  for idx, sentence in tqdm(enumerate(sentences)):
    batch_prompt += sentence + "\n"
    orig_batch.append(sentence)
    if idx % batch_size == 0 and idx > 0:
      batch_prompt = prompt.format(batch_prompt)
      response = model.generate_content(batch_prompt).text
      for translation, original in zip(response.split("\n"), orig_batch):
        json.dump({"Original": original, "Translation": translation.strip()}, f, ensure_ascii=False)
        f.write('\n')
      batch_prompt = ""
      orig_batch = []
      sleep(5)
```
