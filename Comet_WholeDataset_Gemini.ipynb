```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Final_Poster_Machine_Translation/final dataset/EN_CAT_Source_Reference_Translation.csv')  # Replace with your actual file name
df.head()
```

```python

data = []
for _, row in df.iterrows():
    data.append({
        "src": row["Source"],
        "mt": row["Automatic Translation"],
        "ref": row["Reference"]
    })
```

```python
from comet import download_model, load_from_checkpoint

# Load the pre-trained COMET model
model_path = download_model("Unbabel/wmt22-comet-da")
model = load_from_checkpoint(model_path)

# Run COMET scoring
comet_scores = model.predict(data, batch_size=8, gpus=0)["scores"]

# Add the scores to your DataFrame
df["comet_score"] = comet_scores
df.head()
```

```python
# Then display the DataFrame

display(df)
```
